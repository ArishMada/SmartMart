```python
import pandas as pd
```

```python
inv = pd.read_csv("inventory1.csv", encoding = "windows-1252")
inv
```

```python
def get_avg():
    avg = []
    
    for index, row in inv.iterrows():
        dec18 = row["December_2018"]
        jan = row["January"]
        feb = row["February"]
        mar = row["March"]
        apr = row["April"]
        may = row["May"]
        jun = row["June"]
        jul = row["July"]
        aug = row["August"]
        sep = row["September"]
        octo = row["October"]
        nov = row["November"]
        dec19 = row["December_2019"]
        
        total = dec18 + jan + feb + mar + apr + may + jun + jul + aug + sep + octo + nov + dec19
        total = total / 13
        avg.append(total)
        
    return avg

avg = get_avg()
```

```python
inv["QuantityPerYear"] = avg
inv.to_csv('inven.csv', index=False)
```
